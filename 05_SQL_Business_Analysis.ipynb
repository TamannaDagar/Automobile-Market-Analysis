Step 1: Create the SQLite Database

```python
import sqlite3
import pandas as pd 

conn=sqlite3.connect("automobile_market.db")
```

Step 2: Load the Cleaned Dataset

```python
df=pd.read_csv('cleaned_vehicle_sales.csv')
df.head()
```

Step 3: Create the SQL Table

```python
df.to_sql(
    "vehicle_sales",
    conn,
    if_exists="replace",
    index=False
)
```

Table is created automatically and all the data has been loaded.

Step 13. Window Functions

1. Rank Vehicles

```python
pd.read_sql("""
            select 
            make,
            model,
            trim,
            sellingprice,
            rank() over(
            order by sellingprice desc) 
            as PriceRank
            from vehicle_sales;""", conn)
```

2. Dense Rank

```python
pd.read_sql("""
            select 
            make,
            model,
            trim,
            sellingprice,
            dense_rank() over(
            order by sellingprice desc) 
            as DenseRank
            from vehicle_sales;""", conn)
```

3. Row Number By Brand

```python
pd.read_sql("""
            select make,
            model,
            trim,
            sellingprice,
            row_number() over (partition by make
            order by sellingprice desc)
            RowNo
            from vehicle_sales;""", conn)
```

4. Top Vehicles Per Brand ( Doubt)

```python
pd.read_sql("""
            with RankedCars as(
            select make,
            model,
            trim,
            sellingprice,
            row_number() over(
            partition by make
            order by sellingprice desc) rn
            from vehicle_sales)
            
            select * from RankedCars
            ;""", conn)
```

Step 14. Case Statements

1. Vehicle Category

```python
pd.read_sql("""
            select sellingprice,
            case
            when sellingprice> 50000 then "Luxury"
            when sellingprice>25000 then "Premium"
            else "Economy"
            end as category
            from vehicle_sales;""", conn)
```

2. Price Status

```python
pd.read_sql("""
            select sellingprice,
            mmr,
            case
            when sellingprice> mmr then "Above MMR"
            when sellingprice < mmr then "Below MMR"
            else "Equal"   
            end as PriceStatus
            from vehicle_sales;""", conn)
```
