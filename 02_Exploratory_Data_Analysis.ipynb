5. Correlation Analysis

```python
corr= df.select_dtypes(include=np.number).corr()
```

Heatmap

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True,
            cmap='coolwarm')
plt.show()
```

Analysis-
1. Red → Positive correlation (both variables increase together)
2. Blue → Negative correlation (one increases while the other decreases)
3. White/Light → Little or no linear relationship


Findings-
1. Selling Price vs MMR = 0.98 
This is the strongest and most important relationship.

2. Year vs Vehicle Age = -0.99 
Almost a perfect negative correlation.

3. Odometer vs Vehicle Age = 0.77 
Older vehicles generally have higher mileage.

4. Odometer vs Selling Price = -0.58 
Higher mileage tends to reduce the selling price.

5. Odometer vs MMR = -0.59
Higher mileage is associated with lower market value.

6. Year vs Selling Price = 0.59 
Newer vehicles generally sell for higher prices.

7. Condition vs Selling Price = 0.31
Vehicles in better condition tend to sell for higher prices, but the relationship is only moderate.

8. Price Difference vs Price Ratio = 0.38
This moderate positive relationship is expected because both features are derived from selling price and MMR.

6. Time Analysis

6.1 monthly sales volumne

```python
df.columns
```

```python
monthly_order=['January','February','March','April','May','June','July','August','September','October','November','December']
```

```python
monthly_sales = (
    df.groupby(['sale_month'])
      .size()
      .reindex(monthly_order)
      .reset_index(name='Vehicles Sold')
      
)

plt.figure(figsize=(12,6))

sns.lineplot(
    data=monthly_sales,
    x='sale_month',
    y='Vehicles Sold',
    marker='o'
)

plt.title("Monthly Vehicle Sales Trend")
plt.xlabel("Month")
plt.ylabel("Vehicles Sold")
```

Insight:

1. Which month has the highest sales?- February
2. Which month has the lowest sales?- april

6.2 Monthly Sales Revenue

```python
monthly_revenue = (
    df.groupby(['sale_month'])['sellingprice']
      .sum()
      .reindex(monthly_order)
      .reset_index()
)
plt.figure(figsize=(10,6))
sns.lineplot(x='sale_month',y='sellingprice', marker='o',data=monthly_revenue)

plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Total revenue')
plt.xticks(rotation=90)
plt.show()
```

6.3 Yealy Sales
