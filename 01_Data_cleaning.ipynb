Filling Missing values

Make, model, trim, body, color, interior -> Fill with mode

```python
for col in ['make', 'model','trim','body', 'color','interior']:
    df[col]= df[col].fillna(df[col].mode()[0])
```

Reason:

1. Missing percentage is very low (<3%).
2. Mode preserves the most common category.
3. Suitable for business analysis.

Condition, odometer, mmr -> fill with median

```python
for col in ['condition','odometer','mmr']:
    df[col]=df[col].fillna(df[col].median())
```

Reason:
1. These are numerical rating.
2. Median is robust to outliers.
3. Better than mean

Vin-> Drop missing values

```python
df= df.dropna(subset=['vin'])
```

Reason:
1. VIN is the vehicle identifier
2. Missing VINs are extremely rare.
3. A vehicle without an identifier is difficult to validate.

Selling price -> drop missing values

```python
df=df.dropna(subset=['sellingprice'])
```

Reason:

1. Selling price is your target business variable.
2. You cannot analyze a sale without its price.

Sale Date-> drop missing values

DATA QUALITY KPIS

1. Completeness

```python
completeness= (1-df.isnull().sum().sum()/ (df.shape[0]*df.shape[1]))*100
print(f"Completeness: {completeness:.2f}%")
```

2. uniqueness

```python
uniqueness= (1-df.duplicated().sum()/ len(df))* 100
print(f"Uniqueness: {uniqueness:.2f}%")
```

3. Missing values

```python
missing= df.isnull().mean()* 100
print(missing)
```

4. Duplicated Rows

```python
duplicate_rows= df.duplicated().sum()
print(duplicate_rows)
```

5. duplicate vins

```python
duplicate_vins= df['vin'].duplicated().sum()
print(duplicate_vins)
```

6. Validity

```python
validity= (df['sellingprice'] >= 0).mean()*100
valid_odometer= (df['odometer'] >= 0).mean()*100
print(validity)
print(valid_odometer)
```

7. Timeliness

```python
timeliness= df['saledate'].notnull().mean()*100
print(timeliness)
```

8. Consistency

```python
columns = ['make', 'body', 'transmission', 'state', 'color', 'interior']
total = df.shape[0] * len(columns)

inconsistent = (
    df[columns]
    .isnull()
    .sum()
    .sum()
)

consistency = (1 - inconsistent/total) * 100

print(consistency)
```

Overall data quality score

```python
overall_score = (
    completeness +
    uniqueness +
    validity +
    consistency +
    timeliness 
) / 5

print(f"Overall Data Quality Score: {overall_score:.2f}%")
```

OUTLIER DETECTION

```python
import seaborn as sns
```

FOR SELLING pRICE

```python
sns.boxplot(df['sellingprice'])
```

step 1 calculate Q1, Q3 and IQR

```python
Q1 = df['sellingprice'].quantile(0.25)
Q3 = df['sellingprice'].quantile(0.75)

IQR = Q3 - Q1

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
```

step 2 calculate the outlier limits

```python
lower_limit = Q1 - 1.5 *IQR
upper_limit= Q3 + 1.5 * IQR
print("Lower Limit:", lower_limit)
print("Upper Limit:",upper_limit)
```

step 3 Find Outliers

```python
outliers= df[(df['sellingprice']< lower_limit)| 
             (df['sellingprice']> upper_limit)]
outliers
```

step 4 count the outliers

```python
print("Number Of outliers:", outliers.shape[0])
```

step 5 percentage calculation

```python
outlier_percentage = (len(outliers) / len(df)) * 100

print(f"Outlier Percentage: {outlier_percentage:.2f}%")
```

```python
df[df['sellingprice'] <= 0]
```

Decision- Keep Them
Reason

In my dataset, the outliers are likely to include:

Luxury cars (Mercedes-Benz, BMW, Porsche, Ferrari, etc.)
Rare or high-value vehicles
Collector vehicles
Nearly new premium models

These are real business observations, not data errors. Removing them would bias your analysis and underrepresent the premium segment of the market.

Data Cleaning- Unexpected character or keyword

```python
for col in df.select_dtypes(include='object').columns:
    try:
        df[col].str.encode('ascii')
    except Exception:
        print(col)
```

For color column

```python
for value in df['color'].dropna().unique():
    try:
        value.encode('ascii')
    except UnicodeEncodeError:
        print(value)
```

For Interior Column

```python
for value in df['interior'].dropna().unique():
    try:
        value.encode('ascii')
    except UnicodeEncodeError:
        print(value)
```

Remove them

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = (
        df[col]
        .astype(str)
        .str.encode('ascii', errors='ignore')
        .str.decode('ascii')
    )
```
